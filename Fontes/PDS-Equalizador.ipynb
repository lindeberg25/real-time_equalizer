```python


import struct
import numpy as np

import matplotlib.pyplot as plt
from scipy.fftpack import fft
from tkinter import TclError
import wave
import pyaudio
from matplotlib.widgets import Slider

from filtros import *

# Apresenta o display em uma janela Tk separada
%matplotlib tk

# constantes
CHUNK = 1024 * 2 * 2   # amostras por frame
RATE = 44100           # amostras por sengundo
```

```python
# cria o  matplotlib figuras e eixos
fig, (ax2, ax3) = plt.subplots(2, figsize=(15, 7))

wf = wave.open('audio/Back_In_Black.wav', 'rb')


p = pyaudio.PyAudio()

#Abre uma objeto sctream para escrever no arquivos wav
# 'output = True'  indica que o som será tocado ao invés de gravar
stream = p.open(format = p.get_format_from_width(wf.getsampwidth()),
                channels = wf.getnchannels(),
                rate = wf.getframerate(),
                output = True)


# variável para plotar
xf = np.linspace(0, RATE, CHUNK)     # frequencia (espectro)

xf_filter= np.linspace(0, RATE, CHUNK)


# cria uma linha semilogx para o espectro 
line_fft, = ax2.semilogx(xf, np.random.rand(CHUNK), '-', lw=2)

# cria uma objeto de linha randômico 
line_filtered, = ax3.semilogx(xf_filter, np.random.rand(CHUNK), '-', lw=2)


# formato o eixo do espectro
ax2.set_xlim(20, RATE / 2)


# formato o eixo do espectro filtrado
ax3.set_xlim(20, RATE / 2)

# qunatidade de frames
frame_count = 0


# Filtros para cada frequencia
h200hz = gerarFiltro(0.004, 0.005, 0.04)
h400hz = gerarFiltro(0.008, 0.010, 0.04)
h600hz = gerarFiltro(0.012, 0.014, 0.04)
h800hz = gerarFiltro(0.017, 0.019, 0.04)
h1000hz = gerarFiltro(0.021, 0.023, 0.04)
h1200hz = gerarFiltro(0.025, 0.03, 0.04)
h1400hz = gerarFiltro(0.031, 0.033, 0.04)
h1600hz = gerarFiltro(0.035, 0.038, 0.04)
h1800hz = gerarFiltro(0.040, 0.043, 0.04)
h2000hz = gerarFiltro(0.044, 0.046, 0.04)

data = wf.readframes(CHUNK)
    
f1=0
slider=False

a0 = 1
f0 = 0
delta_f = 200
delta_a = 1
    
axcolor = 'lightgoldenrodyellow'
axfreq = plt.axes([0.1, 0.02, 0.55, 0.02], facecolor=axcolor)
axamp = plt.axes([0.1, 0.9, 0.55, 0.02], facecolor=axcolor)

sfreq = Slider(axfreq, 'Freq', 0, 2000, valinit=f0, valstep=delta_f)
samp = Slider(axamp, 'Ganho', 0, 12.0, valinit=a0, valstep=delta_a)

filtro_escolhida= 0

inicia_filtragem=False

# variaveis que recebe "pedaços" do sinal filtrado 
filtro_200hz=0
filtro_400hz=0
filtro_600hz=0
filtro_800hz=0
filtro_1000hz=0
filtro_1200hz=0
filtro_1400hz=0
filtro_1600hz=0
filtro_1800hz=0
filtro_2000hz=0

#variável que recebe o ganho escolhido pelo usuario.
# Elas começam setadas com 1, entretanto, este é o valor padrão e não há ganho. 
# O 1 serve apenas para indicar que o filtro será considerado
ganho_200hz=1
ganho_400hz=1
ganho_600hz=1
ganho_800hz=1
ganho_1000hz=1
ganho_1200hz=1
ganho_1400hz=1
ganho_1600hz=1
ganho_1800hz=1
ganho_2000hz=1

# passa por toda música, lendo em pedaçs (chunck)
while len(data) > 0:


    def update(val):
        ganho = int(samp.val)
        freq = int(sfreq.val)
        global inicia_filtragem
        inicia_filtragem=True
        
        global filtro_200hz
        global filtro_400hz
        global filtro_600hz
        global filtro_800hz
        global filtro_1000hz
        global filtro_1200hz
        global filtro_1400hz
        global filtro_1600hz
        global filtro_1800hz
        global filtro_2000hz
        
        
        global ganho_200hz
        global ganho_400hz
        global ganho_600hz
        global ganho_800hz
        global ganho_1000hz
        global ganho_1200hz
        global ganho_1400hz
        global ganho_1600hz
        global ganho_1800hz
        global ganho_2000hz
        
        if freq == 200:
            filtro_200hz=h200hz
            ganho_200hz=ganho
            print("Filtrando em 200Hz")
            print("Ganho em 200Hz: ", ganho_200hz)
        elif freq == 400:
            filtro_400hz=h400hz
            ganho_400hz=ganho
            print("Filtrando em 400Hz")
            print("Ganho em 400Hz: ", ganho_400hz)
        elif freq == 600:
            filtro_600hz=h600hz
            ganho_600hz=ganho
            print("Filtrando em 600Hz")
            print("Ganho em 600Hz: ", ganho_600hz)
        elif freq == 800:
            filtro_800hz=h800hz
            ganho_800hz=ganho
            print("Filtrando em 800Hz")
            print("Ganho em 800Hz: ", ganho_800hz)
        elif freq == 1000:
            filtro_1000hz=h1000hz
            ganho_1000hz=ganho
            print("Filtrando em 1000Hz")
            print("Ganho em 1000Hz: ", ganho_1000hz)
        elif freq == 1200:
            filtro_1200hz=h1200hz
            ganho_1200hz=ganho
            print("Filtrando em 1200Hz")
            print("Ganho em 1200Hz: ", ganho_1200hz)
        elif freq == 1400:
            filtro_1400hz=h1400hz
            ganho_1400hz=ganho
            print("Filtrando em 1400Hz")
            print("Ganho em 1400Hz: ", ganho_1400hz)
        elif freq == 1600:
            filtro_1600hz=h1600hz
            ganho_1600hz=ganho
            print("Filtrando em 1600Hz")
            print("Ganho em 1600Hz: ", ganho_1600hz)
        elif freq == 1800:
            filtro_1800hz=h1800hz
            ganho_1800hz=ganho
            print("Filtrando em 1800Hz")
            print("Ganho em 1800Hz: ", ganho_1800hz)
        elif freq == 2000:
            filtro_2000hz=h2000hz
            ganho_2000hz=ganho
            print("Filtrando em 2000Hz")
            print("Ganho em 2000Hz: ", ganho_2000hz)
        
    sfreq.on_changed(update)
    samp.on_changed(update)

       
    data_int = struct.unpack('h' * CHUNK  , data)


    # compute FFT and update line
    yf = fft(data_int)
    line_fft.set_ydata(np.abs(yf[0:CHUNK])  / (1024 * CHUNK ))

        
    y_conv200hz = np.convolve(data_int, filtro_200hz).astype('int')
    y_conv400hz = np.convolve(data_int, filtro_400hz).astype('int')
    y_conv600hz = np.convolve(data_int, filtro_600hz).astype('int')
    y_conv800hz = np.convolve(data_int, filtro_800hz).astype('int')
    y_conv1000hz = np.convolve(data_int, filtro_1000hz).astype('int')
    y_conv1200hz = np.convolve(data_int, filtro_1200hz).astype('int')
    y_conv1400hz = np.convolve(data_int, filtro_1400hz).astype('int')
    y_conv1600hz = np.convolve(data_int, filtro_1600hz).astype('int')
    y_conv1800hz = np.convolve(data_int, filtro_1800hz).astype('int')
    y_conv2000hz = np.convolve(data_int, filtro_2000hz).astype('int')
    
    
    # soma as convoluções de cada frequência com seus respectivos ganhos
    # Esta expressão permite filtrar o sinal usando vários filtros simultaneamente
    y_conv = int(ganho_200hz)*y_conv200hz[0:CHUNK] + int(ganho_400hz)*y_conv400hz[0:CHUNK] + int(ganho_600hz)*y_conv600hz[0:CHUNK] + int(ganho_800hz)*y_conv800hz[0:CHUNK] + int(ganho_1000hz)*y_conv1000hz[0:CHUNK] + int(ganho_1200hz)*y_conv1200hz[0:CHUNK] + int(ganho_1400hz)*y_conv1400hz[0:CHUNK] + int(ganho_1600hz)*y_conv1600hz[0:CHUNK] + int(ganho_1800hz)*y_conv1800hz[0:CHUNK] + int(ganho_2000hz)*y_conv2000hz[0:CHUNK]
   
    
    yfiltered=fft(y_conv)
    line_filtered.set_ydata(np.abs(yfiltered[0:CHUNK])  / (256 *CHUNK))
    output_value = y_conv[0:CHUNK]
    
    # variável utilizada para saber quando o processo de filtragem foi iniciada pelo usuário
    if (inicia_filtragem==False):
        output_value = data_int

    output_string = struct.pack('h' * CHUNK, *output_value)

    stream.write(output_string)

                
    data = wf.readframes(CHUNK)
    try:
        fig.canvas.draw()
        fig.canvas.flush_events()
        frame_count += 1

    except TclError:
        print('Terminou !')
        break
    
```
